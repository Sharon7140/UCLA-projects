```python
base_url='https://www.bea.gov'
url='https://www.bea.gov/news/current-releases'
```

```python
def lovely_soup(u):
    page=requests.get(u)
    return(BeautifulSoup(page.content,'html.parser'))
```

```python
soup=lovely_soup(url)
```

```python
soup.findAll('a',href=True)
```

```python
url_list=[]
for a in soup.findAll('a',href=True):
    if a['href'] [0:5]=='/news' :
        url_list.append(base_url+a['href'])
```

```python
import pandas as pd
table_list=[]
url_list_new=url_list[8:11]
url_list_new.append(url_list[13])
url_list_new.append(url_list[15])
url_list_new.append(url_list[20])
url_list_new.append(url_list[24])
url_list_new.append(url_list[26])

for url in url_list_new:
    html=requests.get(url).content
    df_list=pd.read_html(html)
    table_list.append(df_list)
```
